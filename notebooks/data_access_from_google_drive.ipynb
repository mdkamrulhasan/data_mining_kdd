## Objective:  

The objective of this notebook is to showcase how to read and write/dump data from Google Drive

```python
import os
import pandas as pd
```

```python
from google.colab import drive
```

# Data loading from GitHub

```python
features = pd.read_csv("https://raw.githubusercontent.com/mdkamrulhasan/data_mining_kdd/main/data/retail/Features-data-set.csv")
```

## Writing data to your Google Drive Space

```python
base_path = 'drive/MyDrive/data'
```

```python
features.to_csv(os.path.join(base_path, "my-features-set.csv"), index=False)
```
